```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
df['Education'].fillna(df['Education'].mode()[0],inplace=True)
df['Dependents'].fillna(df['Dependents'].mode()[0],inplace=True)
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0],inplace=True)
```

```python
df['ApplicantIncome'].fillna(df['ApplicantIncome'].mean(),inplace=True)
df['CoapplicantIncome'].fillna(df['CoapplicantIncome'].mean(),inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].mean(),inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean(),inplace=True)
df['Credit_History'].fillna(df['Credit_History'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Total_Income']=df['ApplicantIncome']+df['CoapplicantIncome']
df.head()
```

```python
cols =['Loan_ID','Unnamed: 0']
df.drop(columns=cols,axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']
for i in cols:
  df[i]=le.fit_transform(df[i])
```

```python
df.info()
```

```python
for i in df.columns:
  IQR=df[i].quantile(0.75)-df[i].quantile(0.25)
  upper_bound=df[i].quantile(0.75)+1.5*IQR
  lower_bound=df[i].quantile(0.25)-1.5*IQR
  df.loc[df[i]>upper_bound,i]=upper_bound
  df.loc[df[i]<lower_bound,i]=lower_bound
```

```python
df.plot(kind = 'box', figsize=(20,10))
```

```python
df['LoanAmount_log'] = np.log(df['LoanAmount']+1)
sns.distplot(df['LoanAmount_log'])
```

```python
df['ApplicantIncome_log']= np.log(df['ApplicantIncome']+1)
sns.distplot(df['ApplicantIncome_log'])
```

```python
df['Loan_Amount_Term_log']= np.log(df['Loan_Amount_Term']+1)
sns.distplot(df['Loan_Amount_Term_log'])
```

```python
df['Total_Income_log']= np.log(df['Total_Income']+1)
sns.distplot(df['Total_Income_log']+1)
```

```python
cols1 = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Total_Income']
df.drop(columns=cols1,axis=1,inplace=True)
```

```python
df.head()
```

```python
X=df.drop(columns=['Loan_Status'],axis=1)
y= df['Loan_Status']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.20 , random_state=42)
```

```python
X_train
```

```python
y
```

```python
model = DecisionTreeClassifier(max_depth = 3)
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(accuracy_score(y_test, y_predict))
confusion_matrix(y_test, y_predict)
```

```python
accuracy = accuracy_score(y_test, y_predict)
print("accuracy of model: ",accuracy*100)
```

```python
score = cross_val_score(model,X,y,cv=5,scoring='precision')
score
np.mean(score*100)
```

```python
df['Loan_Status'].value_counts()
```

```python
report= classification_report(y_test, y_predict)
print(f"Classification Report or model :\n{report}\n")
```
